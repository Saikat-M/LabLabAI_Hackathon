# Importing all the libraries

```python
import whisper
import gradio as gr 
import openai
from IPython.display import Audio, display
import googletrans  
from googletrans import Translator
import os
from google.cloud import texttospeech_v1
```

#Running an Example audio

```python
display(Audio('/tmp/Recording_2.m4a', autoplay=True))
```

#Downloading whisper model 

```python
model = whisper.load_model('medium')
```

#Transcribing the audio

```python
#Translate Function: Translating the content of the audio to english. 
# The audio is a question asking Who is India's current Prime mininster in BENGALI
translate_output = model.transcribe("/tmp/Recording_2.m4a", task = 'translate')
print(translate_output)
print(f"The Sentence is {translate_output['text']}")
print(f"The language is {translate_output['language']}")
original_audio_language = translate_output['language']
```

#Get Inference from ChatGPT

```python
file = open('/tmp/OpenAPI_API_Key.txt','r')
API_KEY = file.read()
openai.api_key = API_KEY
chat = openai.ChatCompletion.create(
        model="gpt-3.5-turbo", messages= [{"role": "system", "content": translate_output['text']}]
    )
print(chat)
print(chat["choices"][0]["message"]["content"])
```

# Translating the answer back to the original language of the audio

```python
print(googletrans.LANGUAGES)
translator = Translator()

#Translate Function: Back to original language which is bangali in this case
text = chat["choices"][0]["message"]["content"]

lang = translator.detect(text).lang
print(lang)

translated_text = translator.translate(text, src=lang, dest= original_audio_language)

print(translated_text)
print(translated_text.text)
```

# Text to Speech Conversion

```python
def text_to_speech(text, original_language):
  os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = r"/tmp/texttospeechconversion-52f850f6d7ad.json"

  print("text", text, "&", "original_language", original_language)

  # Instantiates a client
  client = texttospeech_v1.TextToSpeechClient()

  # Set the text input to be synthesized
  synthesis_input = texttospeech_v1.SynthesisInput(text=text)


  voice = texttospeech_v1.VoiceSelectionParams(
      language_code=original_language, ssml_gender=texttospeech_v1.SsmlVoiceGender.NEUTRAL
  )

  # Select the type of audio file you want returned
  audio_config = texttospeech_v1.AudioConfig(
      # https://cloud.google.com/text-to-speech/docs/reference/rpc/google.cloud.texttospeech.v1#audioencoding
      audio_encoding=texttospeech_v1.AudioEncoding.MP3
  )

  # Perform the text-to-speech request on the text input with the selected
  # voice parameters and audio file type
  response = client.synthesize_speech(
      input=synthesis_input, voice=voice, audio_config=audio_config
  )

  # The response's audio_content is binary.
  file_name = "Translated.mp3"
  folder = "/tmp"
  file_path = folder+'/'+file_name
  print("file_path: ", file_path)
  with open(file_path, "wb") as out:
      # Write the response to the output file.
      out.write(response.audio_content)
      print("Audio content written to file" + file_name + "in" + folder)

  return file_path
```

#Function handle the user input and generate out from UI based on user preference

```python
def orig_voice_chat(user_voice, Servcie):

    messages = [
    {"role": "system", "content": "You are a kind helpful assistant."},
    ]
    print("Chosen Option", Servcie)

    #translate the message to English
    translate_output = model.transcribe(user_voice, task = 'translate')
    print("translate_output from orig_voice_chat: ",translate_output)    

    user_message = translate_output["text"]
    original_language = translate_output['language']

    if 'Translate' in Servcie:
      user_message = Servcie+translate_output["text"]
    
    #reply = user_message
    print("user_message: ", user_message)
    messages.append(
        {"role": "user", "content": user_message},
    )

    print("messages from orig_voice_chat: ",messages)

    chat = openai.ChatCompletion.create(
        model="gpt-3.5-turbo", messages=messages
    )
    
    reply = chat.choices[0].message.content
    
    print("chatGPT reply in orig_voice_chat", reply)

    messages.append({"role": "assistant", "content": reply})

    # Covert back to original language 
    # tts.tts_to_file(text=reply, file_path="/Translated.mp3")
    lang = translator.detect(reply).lang
    print(lang)

    bengali_translated_text = translator.translate(reply, src=lang, dest= original_language)
    print("bengali_translated_text from orig_voice_chat: ",bengali_translated_text)
    print("bengali_translated_text from orig_voice_chat: ",bengali_translated_text.text)

    file_path = text_to_speech(bengali_translated_text.text, original_language)

    return(reply, file_path)
```

#UI to take user input and provide Output using Gradio

```python
# UI Creation
text_reply = gr.Textbox(label="ChatGPT Text")
voice_reply = gr.Audio('/tmp/Translated.mp3')

gr.Interface(
    title = 'AI Voice Assistant with ChatGPT AI', 
    fn=orig_voice_chat, 
    inputs=[
        gr.inputs.Audio(source="microphone", type="filepath"),
        gr.inputs.Dropdown(
            ["Translate in Bengali", "Translate in Hindi", "Translate in English", "Talk to ChatGPT as well"], label="Servcies"),
    ],

    outputs=[
        text_reply,  voice_reply
    ]).launch(debug = True)
```
